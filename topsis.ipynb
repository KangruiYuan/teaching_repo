```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./data/topsis_score.csv")
data
```

```python
data["argue"] = data["argue"].max() - data["argue"]
data
```

```python
def normalize(array):
    return array / np.sqrt((array **2).sum())

data["score"] = normalize(data["score"])
data["argue"] = normalize(data["argue"])
data
```

```python
Z_plus = data[["score", "argue"]].max()
Z_sub = data[["score", "argue"]].min()
Z_plus, Z_sub
```

```python
def cal_D_plus(series):
    return np.sqrt(((series[["score", "argue"]] - Z_plus) ** 2).sum())


def cal_D_sub(series):
    return np.sqrt(((series[["score", "argue"]] - Z_sub) ** 2).sum())


data["D+"] = data.apply(cal_D_plus, axis=1)
data["D-"] = data.apply(cal_D_sub, axis=1)
data
```

```python
data["S"] = data.apply(lambda x: x["D-"] / (x["D+"] + x["D-"]), axis=1)
data
```

```python
data["S_norm"] = data["S"] / data["S"].sum()
data["rank"] = (-data["S_norm"]).rank()
data
```
